```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
```

```python
plt.bar(x, y)
plt.show()
```

```python
x = ['banana', 'mango']
y = [400, 550]

plt.bar(x, y)
plt.title('Fruits charts')
plt.xlabel('Fruits')
plt.ylabel('Quantity')

plt.show()
```

```python
x = ['Banana', 'Mango', 'Guava', 'WaterMelon', 'Sapota', 'Apple', 'PineApple']
y = [400, 550, 350, 650, 120, 250, 300]

plt.barh(x, y)
plt.title('Fruits charts')
plt.ylabel('Fruits')
plt.xlabel('Quantity')

plt.show();
```

```python
x = ['Banana', 'Mango', 'Guava', 'WaterMelon', 'Sapota', 'Apple', 'PineApple']
y = [400, 550, 350, 650, 120, 250, 300]

plt.bar(x, y, color ='green')
plt.title('Fruits charts')
plt.ylabel('Fruits')
plt.xlabel('Quantity')

plt.show();
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color='blue')
plt.show()
```

```python
## Histograms

x = np.random.normal(170,10,250)
print(x)
```

```python
plt.hist(x)
plt.show()
```

```python
## Pie Charts

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.title('Pie Chart')
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels=mylabels)
plt.title('Pie chart')
plt.show()
```

```python
plt.pie(y, labels=mylabels, startangle=90)
plt.title('Pie chart')
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels=mylabels, explode=myexplode)
plt.show()
```

```python
plt.pie(y, labels=mylabels, explode=myexplode, shadow=True)
plt.show()
```

```python
plt.pie(y, labels=mylabels, explode=myexplode)
plt.legend()
plt.show()
```

```python
plt.pie(y, labels=mylabels, explode=myexplode)
plt.legend(title='Fruits')
plt.show()
```
