```python
import numpy as np

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = np.mean(speed)

print(x)
```

```python
x = np.median(speed)

print(x)
```

```python
from scipy import stats

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
x = stats.mode(speed)

print(x)
```

## Variance

```python
speed = [32,111,138,28,59,77,97]

var = np.var(speed)

print(var)
```

```python
x = np.sqrt(var)
print(x)
```

## Percentiles

```python
import numpy as np

ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]

x = np.percentile(ages, 75)
print(x)
```

```python
y = np.percentile(ages, 25)
print(y)
```

```python
z = np.percentile(ages, 90)
print(z)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.uniform(0, 5, 100000)

plt.hist(x, 100)
plt.show()
```

## Normal Data Distribution

```python
x = np.random.normal(5, 1, 100000)
plt.hist(x, 100)
plt.show()
```
